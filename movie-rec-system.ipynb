```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.rename(columns={'movie_id_x': 'id', 'cast_x': 'cast', 'crew_x': 'crew'}, inplace=True)
print(movies)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()

```

```python
def actlist(obj):
    k=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            k.append(i['character'])
            counter+=1
        else :
            break
    return k
```

```python
movies['cast']= movies['cast'].apply(actlist)
```

```python
movies.head()
```

```python
def directorlist(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
        
```

```python
movies['crew']=movies['crew'].apply(directorlist)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python

vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python



def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
```

```python




new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python





def recommend(movie):
    if movie not in new_df['title'].values:
        print(f"Error: '{movie}' not found in dataset.")
        return
    
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]


    movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])
    
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
